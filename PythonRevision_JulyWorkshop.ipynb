## Dictionaries and Sets

```python
dics = {
'Name' : 'Python',
'Image' : 'Dog',
'set': {1,4,5},
'list' : [1,4,5]
}
#Dictionary Methods
print(dics.items())
print(dics.keys())
print(dics.update({'Image' : 'cat'})) # Updates the values
#Displaying....
print(dics) # This will display all key pairs
print(dics["Image"]) # This will display given key's pair
```

```python
#set =set() empty set declaration
set = {12, 4 ,3 ,9} # Set declaration
print(len(set)) # length of set
#Operations on set
set.remove(12)  # removes 12 from set
print(set)
set.add(8)      # add 8 to set
print(set)
set.pop()       #pop any arbitrary element
print(set)
set.clear()     #clear the set

# set[1] = 2    # It cannot be overwritten
# print[set[2]] # we cannot access a set using its index
```
