# Logistic Regression

```python
import pandas as pd
#col_names = ['pregnancies', 'glucose', 'bp', 'skin', 'insulin', 'bmi', 'pedigree', 'age', 'label']
# load dataset
pima = pd.read_csv("diabetes.csv")
```

```python
pima.head()
```

```python
feature_cols = ['Pregnancies', 'Glucose', 'BloodPressure', 'SkinThickness','Insulin','BMI','DiabetesPedigreeFunction','Age']
X = pima[feature_cols] # Features
y = pima.Outcome # Target variable
```

```python
type(X)
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.25,random_state=0)
```

```python
from sklearn.linear_model import LogisticRegression
logreg = LogisticRegression(max_iter=10000)
```

```python
logreg.fit(X_train,y_train)
```

```python
y_pred=logreg.predict(X_test)
```

```python
from sklearn import metrics
print("Accuracy:",metrics.accuracy_score(y_test, y_pred))
print("Precision:",metrics.precision_score(y_test, y_pred))
print("Recall:",metrics.recall_score(y_test, y_pred))
#y train and y pred should not be compared but then the number of data doesnt match aslwee(unequal)
#beyond 400 any huge value of max iter gives 79 only
#beyond a certain point the accuracy is stagnant
```
